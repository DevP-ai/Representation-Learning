```python
import numpy as np
```

```python
def PCA(x,num_comp):
    #step-1
    x_mean=x-np.mean(x,axis=0)
    
    #step-2
    cov_mat=np.cov(x_mean,rowvar=False)
    
    #step-3
    eig_values,eig_vectors=np.linalg.eigh(cov_mat)
    
    #step-4
    sort_index=np.argsort(eig_values)[::-1]
    sort_eigValues=eig_values[sort_index]
    sort_eigVectors=eig_vectors[:,sort_index]
    
    #step-5
    eigVector_subset=sort_eigVectors[:,0:num_comp]
    
    #step-6
    
    x_reduced=np.dot(eigVector_subset.transpose(),x_mean.transpose()).transpose()
    
    return x_reduced
    
    
```

```python
import pandas as pd
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
```

```python
#Apply it to PCA Function
mat_reduced=PCA(x,2)
```

```python
#Creating a pandas DataFrame of reduced Dataset

principal_df=pd.DataFrame(mat_reduced,columns=['PC1','PC2'])
```
