
**Subtract the mean of each variable
Subtract the mean of each variable from the dataset so 
that the dataset should be centered on the origin. 
Doing this proves to be very helpful when calculating 
the covariance matrix.**



```python
#Genearate a dummy dataset.

data=np.random.randint(10,50,100).reshape(20,5)
# data
```

```python
his=sns.distplot(data)
plt.show()
```

```python
#Centering the mean

data_mean=data-np.mean(data,axis=0)
# data_mean
```

```python
sns.distplot(data_mean)
plt.show()
```

```python
sns.distplot(data,hist=False,color='g')
sns.distplot(data_mean,hist=False,color='r')

plt.show()
```

# ***Calculate the Covariance Matrix***

**The covariance matrix is a square matrix denoting the covariance of the elements with each other. The covariance of an element with itself is nothing but just its Variance.**

```python
#calculating the covariance matrix of the mean-centered data
#rowvar is set False to get the covariance matrix in the required dimensions.


cov_mat=np.cov(data_mean,rowvar=False)
# cov_mat
```

```python
sns.distplot(cov_mat,hist=False,color='r')
plt.show()
```

# **Compute the Eigenvalues and Eigenvectors**

**Now, compute the Eigenvalues and Eigenvectors for the calculated Covariance matrix. The Eigenvectors of the Covariance matrix we get are Orthogonal to each other and each vector represents a principal axis.**

# *A Higher Eigenvalue corresponds to a higher variability. Hence the principal axis with the higher Eigenvalue will be an axis capturing higher variability in the data.*

**Orthogonal means the vectors are mutually perpendicular to each other. Eigenvalues and vectors seem to be very scary until we get the idea and concepts behind it.**


```python
eig_values,eig_vectors=np.linalg.eigh(cov_mat)
```

```python
sns.distplot(eig_values,hist=False,color='r')
sns.distplot(eig_vectors,hist=False,color='k')
plt.show()
```

# ***Sort Eigenvalues in descending order***

Sort the Eigenvalues in the descending order along with their corresponding Eigenvector.

Remember each column in the Eigen vector-matrix corresponds to a principal component, so arranging them in descending order of their Eigenvalue will automatically arrange the principal component in descending order of their variability.

```python
#sort the eignvalues in descending order

sort_index=np.argsort(eig_values)[::-1]
sort_index
```

```python
sort_eig_value=eig_values[sort_index]
sort_eig_value
```

```python
#similarly sort the eigenvactor

sort_eig_vector=eig_vectors[:,sort_index]
sort_eig_vector
```

# ***Select a subset from the rearranged Eigenvalue matrix***

Select a subset from the 
rearranged Eigenvalue matrix as per 
our need i.e. number_comp = 2. 
This means we selected the first 
two principal components

```python
# select the first n eigenvectors, n is desired dimension
# of our final reduced data.

n_comp=3
#we can select any number
eig_vector_subset=sort_eig_vector[:,0:n_comp]
eig_vector_subset
```

# **Transform the data**

Finally, transform the data by having a dot product between the Transpose of the Eigenvector subset and the Transpose of the mean-centered data. By transposing the outcome of the dot product, the result we get is the data reduced to lower dimensions from higher dimensions.

```python
reduce_data=np.dot(eig_vector_subset.transpose(),data_mean.transpose())
reduce_data
```

```python
reduce_data=np.dot(eig_vector_subset.transpose(),data_mean.transpose()).transpose()
reduce_data
```
