# **Removing HTML Tags**

```python
sample_text='<h1>Kaggle</h1><p>Grow your data<code class="w3-codespan">&lt;sup&gt;</code> science skills by competing in our exciting competitions. Find help in the documentation or learn about Community Competitions.</p>'
```

```python
sample_text
```

```python
import re

def striphtml(data):
    p=re.compile(r'<.*?>')
    return p.sub('',data)
```

```python
striphtml(sample_text)
```

# **Tokenization**

```python
dummy="""The institute was established by the International Institute of Information Technology University Act,2013 of the Government of Chhattisgarh, is a joint venture of Government of Chhattisgarh and NTPC Ltd. It is a State University recognized by University Grants Commission, New Delhi.IIIT-NR currently runs several undergraduate programs for B.Tech. degree, Postgraduate M. Tech and MS(R) programmes. In addition, it also offers Ph.D programs in various fields."""
```

```python
dummy
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize
```

```python
sentence=sent_tokenize(dummy)
sentence
```

```python
for sent in sentence:
    print(word_tokenize(sent))
```
